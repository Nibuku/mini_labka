```python
description=list(df.description)
description
```

```python
time=list()
for dict in description:
    time.append(dict[:11])
df["time"]=time
df['description'] = df['description'].str.slice(24)
df
```

```python
teacher=list()
remains=list()
for dict in df.description:
    teacher.append(dict.split('<')[0])
    remains.append(dict.split('<strong>')[1])
df["teacher"]=teacher
df = df.drop('description', axis=1)
df["description"]=remains

df
```

```python
type=list()
groups=list()
for dict in df.description:
    type.append(dict.split('<')[0])
    groups.append(dict.split('<br>')[1])
df["type_of_pair"]=type
df = df.drop('description', axis=1)
df["groups"]=groups
df
```

```python
df['start'] = df['start'].str[:-9]
df=df.drop('end', axis=1)
df=df.rename(columns={"start":"data"})
df
```

```python
dframe = pd.read_csv("tabletime.csv")
```

```python
def data_filter(
    dframe: pd.DataFrame, data:str, type:str
) -> pd.DataFrame:

    filtered_df = dframe[
        (dframe["data"] == data)
        & (dframe["type_of_pair"] == type)
    ]
    return filtered_df

data_filter(dframe, "2024-02-03", "Лабораторные работы")
```

```python
def lesson_filter(dframe: pd.DataFrame, lesson:str):
    filtered_df = dframe[(dframe["title"] == lesson)]
    return filtered_df

lesson_filter(dframe, "Операционные системы")
```

```python
def teacher_filter(dframe: pd.DataFrame, name:str):

    filtered_df = df[df['teacher'].str.contains(name)]
    return filtered_df


teacher_filter(dframe, "Волков Никита Андреевич")
```
